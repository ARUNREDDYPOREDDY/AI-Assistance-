# **QUESTION-1**

**Prompt:** Write a Python program that checks whether a given number is palindrome.

# **CODE**

```python
def is_palindrome(n):
    return str(n) == str(n)[::-1]
```

# **OUTPUT**

```python
user_input = input("Enter a number: ")

try:
    number = int(user_input)
    if is_palindrome(number):
        print(f"{number} is a palindrome.")
    else:
        print(f"{number} is not a palindrome.")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

**EXPLANATION**

This code uses ipywidgets to create an interactive input field for a number. It dynamically checks if the entered number is a palindrome using a dedicated function. The results, including handling invalid inputs, are displayed in real-time below the input box, offering a user-friendly experience

**QUESTION-2**

**Prompt:** Write a Python program to calculate factorial of a given number.

**CODE**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result
```

**OUTPUT**

```python
user_input = input("Enter an integer to calculate its factorial: ")

try:
    number = int(user_input)
    if number < 0:
        print(factorial(number))
    else:
        print(f"The factorial of {number} is: {factorial(number)}")
except ValueError:
    print("Invalid input. Please enter an integer.")
```

**EXPLANATION**

The factorial function handles negative numbers, returning an error message. For 0 or 1, it returns 1 (base cases). For numbers greater than 1, it iteratively multiplies integers from 2 up to the given number. The code also includes user input and error handling.

**QUESTION-3**

**Prompt:** Write a Python function to check Armstrong number.

**CODE**

```python
def is_armstrong(n):
    digits = str(n)
    power = len(digits)
    total = 0

    for d in digits:
        total += int(d) ** power

    if total == n:
        return f"{n} is an Armstrong Number"
    else:
        return f"{n} is not an Armstrong Number"
```

**OUTPUT**

```python
user_input = input("Enter a number to check if it's an Armstrong number: ")

try:
    number = int(user_input)
    if number < 0:
        print("Armstrong numbers are typically defined for positive integers.")
    else:
        print(is_armstrong(number))
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### **Explanation**

This function converts the input number to a string to count its digits. It then calculates the sum of each digit raised to the power of the total number of digits. Finally, it compares this sum with the original number to determine if it's an Armstrong number. The code also includes user input and basic error handling for non-integer inputs.

**QUESTION-4**

**Prompt:** Write an optimized Python program to classify a number as Prime, Composite, or Neither.
Validate input.

**CODE**

```python
import math

def classify_number(num):
    if not isinstance(num, int):
        return "Invalid input: Please provide an integer."

    if num < 0:
        return f"{num} is Neither Prime nor Composite (negative numbers)."
    elif num == 0 or num == 1:
        return f"{num} is Neither Prime nor Composite."
    elif num == 2:
        return f"{num} is a Prime number."
    elif num % 2 == 0:
        return f"{num} is a Composite number."
    else:
        for i in range(3, int(math.sqrt(num)) + 1, 2):
            if num % i == 0:
                return f"{num} is a Composite number."
        return f"{num} is a Prime number."

```

**OUTPUT**

```python
user_input = input("Enter an integer to classify: ")

try:
    number = int(user_input)
    print(classify_number(number))
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### **Explanation**

This optimized program classifies an integer as Prime, Composite, or Neither. It first handles negative numbers, 0, and 1 as 'Neither'. For numbers greater than 1, it efficiently checks for divisibility up to its square root, skipping even numbers. If no divisors are found, it's Prime; otherwise, it's Composite.

**QUESTION-5**

**Prompt:** Write a Python function to check whether a number is a perfect number.

**CODE**

```python
def is_perfect_number(num):
    if not isinstance(num, int) or num <= 0:
        return f"{num} is not a valid input for a perfect number (must be a positive integer)."

    # Find all proper divisors (divisors excluding the number itself)
    divisors_sum = 0
    for i in range(1, num):
        if num % i == 0:
            divisors_sum += i

    # A number is perfect if the sum of its proper divisors equals the number itself
    if divisors_sum == num:
        return f"{num} is a Perfect Number."
    else:
        return f"{num} is not a Perfect Number."
```

**OUTPUT**

```python
user_input = input("Enter a positive integer to check if it's a perfect number: ")

try:
    number = int(user_input)
    print(is_perfect_number(number))
except ValueError:
    print("Invalid input. Please enter an integer.")
```

**Explanation**

This function checks if a positive integer is a perfect number. It calculates the sum of all proper divisors (excluding the number itself). If this sum equals the original number, it's classified as perfect. The code includes input validation for positive integers and handles non-integer inputs gracefully. Examples of perfect numbers are 6, 28, 496, and 8128.

**QUESTION-10**

**Prompt:** Write a Python program to find if a number is Even or Odd, by user input and provide an explanation.

**CODE**

```python
def check_even_odd(num):
    if not isinstance(num, int):
        return "Invalid input: Please provide an integer."

    if num % 2 == 0:
        return f"{num} is an Even number."
    else:
        return f"{num} is an Odd number."
```

**OUTPUT**

```python
user_input = input("Enter an integer to check if it's Even or Odd: ")

try:
    number = int(user_input)
    print(check_even_odd(number))
except ValueError:
    print("Invalid input. Please enter an integer.")
```

### **Explanation**

This Python program defines a function `check_even_odd` that determines if a given integer is even or odd. It first validates the input to ensure it's an integer. Then, it uses the modulo operator (`%`) to check if the number has a remainder of 0 when divided by 2. If it does, the number is even; otherwise, it's odd. The program also includes user input and error handling for non-integer inputs.
